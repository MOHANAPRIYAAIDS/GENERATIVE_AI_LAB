```python
def clean_job_skills(job_df):
    job_df["Skill"] = job_df["Skill"].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else [])
    job_df["Skill"] = job_df["Skill"].apply(lambda x: [s.lower() for s in x])
    return job_df

job_df = clean_job_skills(job_df)

def build_skill_vocab(job_df):
    skills = set()
    for s in job_df["Skill"]:
        skills.update(s)
    return list(skills)

skill_vocab = build_skill_vocab(job_df)
```

```python
def resume_analysis(resume_text, role):
    prompt = f"""
    You are an expert resume reviewer.

    Analyze this resume for the role of {role}.

    Evaluate:
    - Skills relevance
    - Project impact
    - ATS optimization

    Output:
    Score (0-100)
    Strengths
    Weaknesses
    Suggestions

    Resume:
    {resume_text}
    """
    return generate_response(prompt)
```

```python
def extract_skills(resume_text):
    text = resume_text.lower()
    return list(set([skill for skill in skill_vocab if skill in text]))
```

```python
def get_job_skills(role):
    filtered = job_df[job_df["Job_Title"].str.lower().str.contains(role.lower())]
    return filtered.iloc[0]["Skill"] if not filtered.empty else []
```

```python
def skill_gap_prompt(user_skills, job_skills, role):
    prompt = f"""
    Compare candidate skills with job requirements.

    Role: {role}

    Candidate Skills:
    {user_skills}

    Required Skills:
    {job_skills}

    Output:
    Missing Skills
    Weak Areas
    Recommendations
    """
    return generate_response(prompt)
```

```python
def recommend_courses(missing_skills):
    matched = course_df[
        course_df["course_title"].str.contains('|'.join(missing_skills), case=False, na=False)
    ]

    courses = matched["course_title"].head(5).tolist()

    prompt = f"""
    Recommend best courses for:
    {missing_skills}

    Available:
    {courses}

    Output:
    Course recommendations with reasons
    """
    return generate_response(prompt)
```

```python
def generate_question(role):
    if role.lower() in ["hr", "manager"]:
        q = random.choice(hr_questions)
        return q["question"], q["ideal_answer"]
    else:
        q = random.choice(ml_questions)
        return q["question"], q["answer"]
```

```python
def evaluate_answer(question, user_answer, ideal_answer):
    prompt = f"""
    Evaluate the answer.

    Question:
    {question}

    User Answer:
    {user_answer}

    Ideal Answer:
    {ideal_answer}

    Criteria:
    - Accuracy
    - Clarity
    - Completeness

    Output:
    Score (1-10)
    Strengths
    Weaknesses
    Improved Answer
    """
    return generate_response(prompt)
```

```python
def learning_plan(missing_skills):
    prompt = f"""
    Create a 2-week study plan.

    Skills:
    {missing_skills}

    Think step-by-step.

    Output:
    Day-wise plan
    Tools
    Mini projects
    """
    return generate_response(prompt)
```

```python
def full_pipeline(resume_text, role, user_answer):

    # Resume
    resume_result = resume_analysis(resume_text, role)

    # Skills
    user_skills = extract_skills(resume_text)
    job_skills = get_job_skills(role)

    # Gap
    gap = skill_gap_prompt(user_skills, job_skills, role)

    # Courses
    courses = recommend_courses(user_skills)

    # Learning plan
    plan = learning_plan(user_skills)

    # Interview
    question, ideal = generate_question(role)

    # Evaluation
    evaluation = evaluate_answer(question, user_answer, ideal)

    return {
        "resume_analysis": resume_result,
        "skills": user_skills,
        "skill_gap": gap,
        "courses": courses,
        "learning_plan": plan,
        "question": question,
        "ideal_answer": ideal,
        "evaluation": evaluation
    }
```

```python
resume_text = resume_df.iloc[0]["text"]
role = "Data Scientist"
answer = "Machine learning is a method where models learn from data."

result = full_pipeline(resume_text, role, answer)

from pprint import pprint
pprint(result)
```

```python
# ==============================
# INSTALL REQUIRED LIBRARIES
# ==============================
!pip install PyPDF2 python-docx transformers sentencepiece --quiet

# ==============================
# IMPORTS
# ==============================
import pandas as pd
import json, random, ast
from transformers import pipeline
from google.colab import files
import PyPDF2, docx

# ==============================
# LOAD MODEL
# ==============================
generator = pipeline("text-generation", model="google/flan-t5-base")

def generate_response(prompt):
    return generator(prompt, max_length=256)[0]['generated_text']

# ==============================
# SAMPLE DATA (Replace with your dataset if needed)
# ==============================
job_df = pd.DataFrame({
    "Job_Title": ["Data Scientist", "ML Engineer"],
    "Skill": [["python","machine learning","sql","deep learning"],
              ["python","pytorch","mlops","deep learning"]]
})

course_df = pd.DataFrame({
    "course_title": [
        "Machine Learning Basics",
        "Deep Learning with PyTorch",
        "SQL for Data Science",
        "MLOps Fundamentals"
    ]
})

ml_questions = [
    {"question": "What is overfitting?", "answer": "Model memorizes training data and fails to generalize."},
    {"question": "Explain gradient descent.", "answer": "Optimization algorithm to minimize loss."}
]

hr_questions = [
    {"question": "Tell me about yourself.", "ideal_answer": "Structured summary of background."}
]

# ==============================
# BUILD SKILL VOCAB
# ==============================
skill_vocab = list(set([s for sublist in job_df["Skill"] for s in sublist]))

# ==============================
# FILE UPLOAD (COLAB)
# ==============================
print("Upload your resume (PDF or DOCX)")
uploaded = files.upload()
file_name = list(uploaded.keys())[0]

# ==============================
# EXTRACT TEXT
# ==============================
def extract_text(file_name):
    text = ""
    if file_name.endswith(".pdf"):
        with open(file_name, "rb") as f:
            reader = PyPDF2.PdfReader(f)
            for page in reader.pages:
                text += page.extract_text() or ""
    elif file_name.endswith(".docx"):
        doc = docx.Document(file_name)
        text = "\n".join([p.text for p in doc.paragraphs])
    return text

resume_text = extract_text(file_name)

print("\n--- RESUME PREVIEW ---\n")
print(resume_text[:500])

# ==============================
# CORE FUNCTIONS
# ==============================
def extract_skills(text):
    text = text.lower()
    return list(set([s for s in skill_vocab if s in text]))

def get_job_skills(role):
    df = job_df[job_df["Job_Title"].str.lower().str.contains(role.lower())]
    return df.iloc[0]["Skill"] if not df.empty else []

def skill_gap(user_skills, job_skills, role):
    prompt = f"""
    Role: {role}
    Candidate Skills: {user_skills}
    Required Skills: {job_skills}

    Identify:
    Missing Skills
    Suggestions
    """
    return generate_response(prompt)

def recommend_courses(missing):
    prompt = f"""
    Suggest courses to learn:
    {missing}
    """
    return generate_response(prompt)

def get_question(role):
    if role.lower() == "hr":
        q = random.choice(hr_questions)
        return q["question"], q["ideal_answer"]
    else:
        q = random.choice(ml_questions)
        return q["question"], q["answer"]

def evaluate_answer(question, user_answer, ideal):
    prompt = f"""
    Question: {question}
    User Answer: {user_answer}
    Ideal Answer: {ideal}

    Give Score (1-10) and improvement
    """
    return generate_response(prompt)

# ==============================
# MAIN EXECUTION
# ==============================
role = input("\nEnter target role (e.g., Data Scientist): ")

# Skills
user_skills = extract_skills(resume_text)
job_skills = get_job_skills(role)
missing = list(set(job_skills) - set(user_skills))

print("\n--- SKILLS ---")
print("Your Skills:", user_skills)
print("Required Skills:", job_skills)

# Gap
print("\n--- SKILL GAP ANALYSIS ---")
print(skill_gap(user_skills, job_skills, role))

# Interview
print("\n--- INTERVIEW ---")
for i in range(2):
    q, ideal = get_question(role)
    print(f"\nQ{i+1}: {q}")
    ans = input("Your Answer: ")
    print("Feedback:", evaluate_answer(q, ans, ideal))

# Courses
print("\n--- COURSE RECOMMENDATIONS ---")
print(recommend_courses(missing))
```
